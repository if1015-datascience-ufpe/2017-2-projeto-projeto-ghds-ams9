Extração de dados via API AppCivico que possui mais atributos, porém nao possui a média do ENEM.

```python
#Escolas de Olinda-PE
codes = rq.get('http://mobile-aceite.tcu.gov.br:80/nossaEscolaRS/rest/escolas?municipio=Olinda&uf=PE&quantidadeDeItens=1000')
data = codes.text
```

```python
jsonEscolas = codes.json()
escolas = []

codEscola = []
nomeEscola = []
tipoEscola = []  #Publica x Privada
situacaoEscola = []   #Em atividade ou Não
qtdSalasEscola = []
qtdFuncionarios = []
qtdComputadores = []
qtdComputadoresPorAluno = []
qtdAlunos = []
zonaEscola = []
#InfraEstrutura da Escola
infraQuadraCoberta = []
infraQuadraDescoberta = []
infraInternet = []
infraLabInformatica = []
infraLabCiencias = []
infraBiblioteca = []

#print(len(escolas))
for x in jsonEscolas:
    codEscola.append(x['codEscola'])
    nomeEscola.append(x['nome'])
    tipoEscola.append(x['rede'])
    situacaoEscola.append(x['situacaoFuncionamento'])
     
    #Verificar conjunto de dados -- Escolas em atividade porém sem informação de quantidade...
  #  if(x['situacaoFuncionamento'] == 'Em Atividade'):
  #      qtdAlunos.append(x['qtdAlunos'])
  #  else:
  #      qtdAlunos.append(0)
  
    qtdSalasEscola.append(x['qtdSalasUtilizadas']) 
    qtdFuncionarios.append(x['qtdFuncionarios'])
    qtdComputadores.append(x['qtdComputadores'])
    qtdComputadoresPorAluno.append(x['qtdComputadoresPorAluno']) 
    zonaEscola.append(x['zona'])
     
    infraQuadraCoberta.append(x['infraestrutura']['temQuadraEsporteCoberta']) 
    infraQuadraDescoberta.append(x['infraestrutura']['temQuadraEsporteDescoberta'])
    infraInternet.append(x['infraestrutura']['temInternet'])
    infraLabInformatica.append(x['infraestrutura']['temLaboratorioInformatica'])
    infraLabCiencias.append(x['infraestrutura']['temLaboratorioCiencias'])
    infraBiblioteca.append(x['infraestrutura']['temBiblioteca'])

myTable = {'CodEscola': codEscola, 'Nome': nomeEscola, 'Tipo' : tipoEscola, 'Situação' : situacaoEscola, 'Nº Salas' : qtdSalasEscola,
           'Nº Funcionarios': qtdFuncionarios, 'Qtd Computadores' : qtdComputadores, 
           'Qtd Aluno/Computador' : qtdComputadoresPorAluno, 'Zona da Escola' : zonaEscola,'Tem Quadra Cob' : infraQuadraCoberta,
          'Tem Quadra Desc': infraQuadraDescoberta, 'Tem Internet' : infraInternet, 'Tem Lab Informática' : infraLabInformatica,
          'Tem Lab Ciências': infraLabCiencias, 'Tem Biblioteca':infraBiblioteca}


dataFrame = pd.DataFrame(myTable)
dataFrame
          #escolas   
#teste = pd.read_json(x, encoding='utf-8')

 
   
    
```

Com os dados acima, podemos perceber que provavelmente, investir na compra de computadores certamente seria mais importante para aumentar a média do Enem de uma determinada escola, do que aumentar a quantidade de salas e funcionários.

Já nas celulas abaixo, iremos utilizar múltiplos atributos e checar as suas respectivas influências nas médias do ENEM. Para tanto vamos converter os atributos do Tipo Sim e Não para 1's e 0's para podermos aplicar os algoritmos de regressão.

```python
from sklearn.preprocessing import LabelEncoder

teste = dataEscolasEnem


attrsToConv = ['Tipo','Tem Biblioteca','Tem Internet', 'Tem Lab Ciências', 'Tem Lab Informática', 'Tem Quadra Cob', 'Tem Quadra Desc'] 
le = LabelEncoder()

for i in attrsToConv: teste[i] = le.fit_transform(teste[i])
teste['MÉDIA GERAL']
teste = teste.rename(index=str, columns={"Tem Biblioteca": "TemBiblioteca", "Tem Internet": "TemInternet", "Tem Lab Ciências": "TemLabCiências", 
                                         "MÉDIA GERAL": "MÉDIAGERAL", "Tem Quadra Cob": "TemQuadraCob", "MÉDIA HUMANAS": "MÉDIAHUMANAS",
                                        "Tem Quadra Desc": "TemQuadraDesc", "Qtd Computadores": "QtdComputadores",
                                        "Nº Funcionarios": "QtdFuncionarios", "Nº Salas": "QtdSalas", "MÉDIA NATUREZA":"MÉDIANATUREZA",
                                        "MÉDIA MATEMATICA": "MÉDIAMATEMATICA"})
teste['MÉDIAGERAL'] = teste['MÉDIAGERAL']/1000
teste['MÉDIANATUREZA'] = teste['MÉDIANATUREZA']/1000
teste['MÉDIAMATEMATICA'] = teste['MÉDIAMATEMATICA']/1000
teste['MÉDIA REDAÇAO'] = teste['MÉDIA REDAÇAO']/1000
teste['MÉDIA LINGUAGENS'] = teste['MÉDIA LINGUAGENS']/1000
teste['MÉDIAHUMANAS'] = teste['MÉDIAHUMANAS']/1000
teste
```

```python

model = smf.logit("MÉDIAGERAL ~ TemQuadraDesc + QtdSalas + QtdFuncionarios + QtdComputadores + TemBiblioteca"
                  , teste)
result = model.fit()
pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

```python
model2 = smf.logit("MÉDIAGERAL ~ QtdSalas + QtdFuncionarios + QtdComputadores + TemLabCiências"
                  , teste)
result = model2.fit()
pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

```python
model3 = smf.logit("MÉDIANATUREZA ~ TemBiblioteca + TemLabCiências + QtdComputadores"
                  , teste)
result = model3.fit()
pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

```python
model4 = smf.logit("MÉDIAMATEMATICA ~ TemQuadraDesc + QtdComputadores + QtdFuncionarios + QtdSalas + TemBiblioteca"
                  , teste)
result = model4.fit()
pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```

Agora vamos buscar observar a importância do tipo da escola

```python
model5 = smf.logit("MÉDIAGERAL ~ Tipo + QtdComputadores + QtdFuncionarios + QtdSalas + TemBiblioteca + TemLabCiências"
                  , teste)
result = model5.fit()
pd.concat([result.params,result.bse,result.tvalues,result.pvalues],
          axis=1, keys=['coef','SE','t','p-value'])
```
